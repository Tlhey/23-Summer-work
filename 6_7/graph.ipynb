# One dimensional graphic
This python program visualize one-dimensional movement, with fixed x[i] position.
The input of the program is a *.csv file.
For every line, it should follow the style: 

```python
T,y[0], y[1], ..., y[N]'\n'
```

Something that might need to be changed:
### N, k, fps, note.set_text

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import openpyxl

dt=0.01
N=100
k="4"
L=1.5
dl=L/N

## read data
data_y = pd.read_csv(k+".csv")
print("data is loaded")
print("shape:")
print(data_y.shape)
## initialize the size
fig, ax = plt.subplots()

ax.set_xlim(-0.1, L+0.1)
ax.set_ylim(-1, 1)

line, = ax.plot([], [], lw = 3)

def init():
  line.set_data([],[])
  return line,



N_ani = ax.text(L-0.3, 0.8,'',fontsize=12)
dt_ani= ax.text(L-0.3, 0.7,'',fontsize=12)
T_ani = ax.text(L-0.3, 0.6,'',fontsize=12)
note  = ax.text(0,     0.8,'',fontsize=12)
def update(frame):
    frame_data_y = data_y.iloc[frame][1:].values
    frame_data_x = np.arange(0,L+dl,dl)
    line.set_data(frame_data_x, frame_data_y)
    N_ani.set_text('X=%d'%N)
    dt_ani.set_text('dt=%.8f'%dt)
    T_ani.set_text('T=%.8f'%data_y.iloc[frame][0])
    note.set_text("init with parabola maxhight=0.3")
    return [line, N_ani, dt_ani, T_ani, note]

ani = FuncAnimation(fig
           ,func=update
           ,init_func=init
           ,frames=len(data_y)
           ,interval=1
           ,blit=True
           )


ani.save(k+".gif", writer = 'pillow', fps = 10)
ani.save(k+".mp4", writer = 'ffmpeg', fps = 10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

#åˆå§‹åŒ–ç”»å¸ƒ
fig = plt.figure()
plt.grid(ls='--')

#ç»˜åˆ¶ä¸€æ¡æ­£å¼¦å‡½æ•°æ›²çº¿
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

crave_ani = plt.plot(x,y,'red',alpha=0.5)[0]

#ç»˜åˆ¶æ›²çº¿ä¸Šçš„åˆ‡ç‚¹
point_ani = plt.plot(0,0,'r',alpha=0.4,marker='o')[0]

#ç»˜åˆ¶xã€yçš„åæ ‡æ ‡è¯†
xtext_ani = plt.text(5,0.8,'',fontsize=12)
ytext_ani = plt.text(5,0.7,'',fontsize=12)
ktext_ani = plt.text(5,0.6,'',fontsize=12)

#è®¡ç®—åˆ‡çº¿çš„å‡½æ•°
def tangent_line(x0,y0,k):
	xs = np.linspace(x0 - 0.5,x0 + 0.5,100)
	ys = y0 + k * (xs - x0)
	return xs,ys

#è®¡ç®—æ–œçŽ‡çš„å‡½æ•°
def slope(x0):
	num_min = np.sin(x0 - 0.05)
	num_max = np.sin(x0 + 0.05)
	k = (num_max - num_min) / 0.1
	return k

#ç»˜åˆ¶åˆ‡çº¿
k = slope(x[0])
xs,ys = tangent_line(x[0],y[0],k)
tangent_ani = plt.plot(xs,ys,c='blue',alpha=0.8)[0]

#æ›´æ–°å‡½æ•°
def updata(num):
	k=slope(x[num])
	xs,ys = tangent_line(x[num],y[num],k)
	tangent_ani.set_data(xs,ys)
	point_ani.set_data(x[num],y[num])
	xtext_ani.set_text('x=%.3f'%x[num])
	ytext_ani.set_text('y=%.3f'%y[num])
	ktext_ani.set_text('k=%.3f'%k)
	return [point_ani,xtext_ani,ytext_ani,tangent_ani,k]

ani = animation.FuncAnimation(fig=fig,func=updata,frames=np.arange(0,100),interval=100)
ani.save('sin_x.gif')
plt.show()

```
